```python
import warnings
warnings.filterwarnings('ignore')

# Visualization settings
plt.style.use('seaborn-v0_8')
plt.rcParams['figure.figsize'] = (14, 7)
plt.rcParams['font.size'] = 11
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.time
```

```python
df['Hour'] = pd.to_datetime(df['Time'], format='%H:%M:%S', errors='coerce').dt.hour
df['Day'] = df['Date'].dt.day_name()
df['Month'] = df['Date'].dt.month
```

```python
df.isnull().sum()
```

```python
df['Driver Ratings'].fillna(df['Driver Ratings'].median(), inplace=True)
df['Customer Rating'].fillna(df['Customer Rating'].median(), inplace=True)

df['Reason for cancelling by Customer'].fillna('Not Applicable', inplace=True)
df['Driver Cancellation Reason'].fillna('Not Applicable', inplace=True)
df['Incomplete Rides Reason'].fillna('Not Applicable', inplace=True)
```

```python
daily_bookings = df.groupby('Date')['Booking ID'].count()

daily_bookings.plot()
plt.title('Daily Uber Bookings Trend')
plt.xlabel('Date')
plt.ylabel('Number of Bookings')
plt.show()
```

```python
import matplotlib.pyplot as plt

(df[df['booking_status'] != 'Completed']
 .groupby('booking_status')['booking_value']
 .sum()
 .sort_values(ascending=False)
 .plot(kind='bar', figsize=(8,5)))

plt.title("Revenue Loss by Ride Failure Type")
plt.xlabel("Ride Failure Type")
plt.ylabel("Total Revenue Lost")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['ride_failure_type'] = np.select(
    [
        df['cancelled_rides_by_customer'] == 1,
        df['cancelled_rides_by_driver'] == 1,
        df['incomplete_rides'] == 1
    ],
    [
        'Cancelled by Customer',
        'Cancelled by Driver',
        'Incomplete Ride'
    ],
    default='Completed'
)
```

```python
(df[df['ride_failure_type'] != 'Completed']
 .groupby('ride_failure_type')['booking_value']
 .sum()
 .sort_values(ascending=False)
 .plot(kind='bar', figsize=(8,5)))

plt.title("Revenue Loss by Ride Failure Type")
plt.xlabel("Ride Failure Type")
plt.ylabel("Total Revenue Lost")
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()
```

```python
df.groupby('ride_failure_type')['booking_value'].describe()
```

```python
avg_completed_value = df[df['booking_status'] == 'Completed']['booking_value'].mean()

df['potential_revenue_loss'] = np.where(
    df['ride_failure_type'] != 'Completed',
    avg_completed_value,
    0
)
```

```python
(df[df['ride_failure_type'] != 'Completed']
 .groupby('ride_failure_type')['potential_revenue_loss']
 .sum()
 .plot(kind='bar', color = 'blue' ,figsize=(8,5)))

plt.title("Potential Revenue Loss by Ride Failure Type")
plt.ylabel("Estimated Revenue Loss")
plt.show()
```
